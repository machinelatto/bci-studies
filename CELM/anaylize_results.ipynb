```python
import numpy as np
import scipy
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Cross Subject Results

### 4 Freqs - 8 10 12 15

```python
results_4_freqs = pd.read_csv(
    "best_trained_raw_models/cross_subject/4freq/results_cnn1d_cross_subj_4_freqs.csv",
    index_col=0,
)
results_4_freqs.loc["mean"] = results_4_freqs.mean(axis=0)
results_4_freqs["accuracy"] = results_4_freqs["accuracy"].apply(
    lambda x: float("{:.4f}".format(x))
)
plt.figure(figsize=(16, 8), dpi=300)
ax = sns.barplot(results_4_freqs["accuracy"] * 100)
last_bar = ax.patches[-1]
last_bar.set_facecolor("red")  # You can change 'red' to any color you
print(ax.containers[0])
ax.bar_label(ax.containers[0])
plt.title("Accuracy for 4 freqs", fontweight="bold", fontsize=20)
plt.xlabel("Subject", fontweight="bold", fontsize=20)
plt.ylabel("Accuracy (%)", fontweight="bold", fontsize=20)
plt.tight_layout()
plt.show()
```
